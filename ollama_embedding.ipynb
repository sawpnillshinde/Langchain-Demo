```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(OllamaEmbeddings(model='gemma:2b'))
```

```python
embeddings
```

```python
response1=embeddings.embed_documents(
     [
       "Alpha is the first letter of Greek alphabet",
       "Beta is the second letter of Greek alphabet", 
    ]                               
)
```

```python
response1
```

```python
text='How many contienents are there in the world?'
response2=embeddings.embed_query(text)
```

```python
response2
```
