```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['OPENAI_API_KEY']=os.getenv('OPENAI_API_KEY')
os.environ['LANGCHAIN_API_KEY']=os.getenv('LANGCHAIN_API_KEY')
os.environ['LANGCHAIN_PROJECT']=os.getenv('LANGCHAIN_PROJECT')
os.environ['LANGCHAIN_TRACING_V2']='true'
```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model='gpt-4o-mini')
print(llm)
```

```python
result=llm.invoke('What is cricket?')
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate(
    [
        ('system','You are ai expert,please provide the answers for the questions'),
        ('user','{input}')
    ]
)

prompt
```

```python
chain=prompt|llm

response=chain.invoke({'input':"What is cricket?Give 5 famous personalities."})

print(response)
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parse=StrOutputParser()

chain=prompt|llm|output_parse
```
